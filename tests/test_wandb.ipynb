```python
import wandb
import torch
import numpy as np

from waggon.utils import display
from waggon.surrogates import DGP
from waggon.acquisitions import EI
from waggon import functions as f
from waggon.optim import SurrogateOptimiser
```

```python
wandb.login()
```

```python
config = {
    'method': 'grid',
    'metric': {
        'name': 'n_steps',
        'goal': 'minimize'
    },
    'parameters': {
        'n_epochs': {
            'distribution': 'categorical',
            'values': [100, 200]
        },
        'lr': {
            'distribution': 'categorical',
            'values': [1e-1, 1e-2]
        },
        'hidden_size': {
            'distribution': 'categorical',
            'values': [32, 128]# 16, 64 ## 256
        },
        'num_inducing': {
            'distribution': 'categorical',
            'values': [16, 64]#32 ## 96
        },
        'n_candidates': {
            'distribution': 'categorical',
            'values': [32, 128]#64
        },
        'actf': {
            'distribution': 'categorical',
            'values': ['tanh']#None, 'relu', 
        },
        'means': {
            'distribution': 'categorical',
            'values': [['linear', 'constant'], ['constant', 'linear'], ['linear', 'linear']]#['constant', 'constant'], 
        },
        'scale': {
            'distribution': 'categorical',
            'values': [True, False]#
        }
    }
}
```

```python
def test():
    
    config_defaults = {
        'n_epochs':      100,
        'lr':            1e-1,
        'hidden_size':   16,
        'num_inducing':  22,
        'n_candidates':  4,
        'actf':          None,
        'means':         ['constant', 'constant'],
        'scale':         True,
    }
    
    wandb.init(config=config_defaults)
    config = wandb.config
    
    if config.actf == 'relu':
        actf = torch.relu
    elif config.actf == 'tanh':
        actf = torch.tanh
    else:
        actf = None
    
    surr = DGP(
        n_epochs     = config.n_epochs,
        lr           = config.lr,
        hidden_size  = config.hidden_size,
        num_inducing = config.num_inducing,
        actf         = actf,
        means        = config.means,
        scale        = config.scale
    )
    
    opt = SurrogateOptimiser(
        func         = f.three_hump_camel(n_obs=1),
        surr         = surr,
        acqf         = EI(log_transform=True),
        seed         = 2,
        error_type   = 'x',
        n_candidates = config.n_candidates
    )
    _ = opt.optimise()
    
    wandb.log({'error': np.round(opt.errors[-1], 5)})
    wandb.log({'n_steps': len(opt.errors)})
```

```python
sweep_id = wandb.sweep(config, project='dgp tuning 032')
```

```python
wandb.agent(sweep_id, function=test)
```
